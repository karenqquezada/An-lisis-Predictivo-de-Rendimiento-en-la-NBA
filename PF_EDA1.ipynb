```python
#calcular puntos por jugador

# Crear copia del DataFrame
pbp = play_by_play_clean.copy()

# Crear columna de puntos por evento
def obtener_puntos(row):
    if row['eventmsgtype'] == 3:      # Tiro libre convertido
        return 1
    elif row['eventmsgtype'] == 1:    # Tiro de campo convertido (asumimos 2)
        return 2
    else:
        return 0

# Aplicar la funci√≥n
pbp['puntos'] = pbp.apply(obtener_puntos, axis=1)

# Agrupar por nombre del jugador y sumar los puntos
top_jugadores = (
    pbp.groupby('player1_name')['puntos']
    .sum()
    .reset_index()
    .sort_values(by='puntos', ascending=False)
    .head(10)
)

# Visualizar
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=top_jugadores, x='puntos', y='player1_name', palette='rocket')

plt.title('üèÄ Top 10 Jugadores con m√°s puntos (eventos)')
plt.xlabel('Puntos')
plt.ylabel('Jugador')

# Mostrar los valores encima de las barras
for i, valor in enumerate(top_jugadores['puntos']):
    ax.text(valor + 1, i, int(valor), color='black', va='center')

plt.tight_layout()
plt.show()
```

Este gr√°fico muestra los diez jugadores con la mayor cantidad de puntos acumulados (o eventos de puntuaci√≥n) en el conjunto de datos analizado.
LeBron James lidera la lista de forma destacada, con 27,057 puntos, estableciendo una clara diferencia sobre el segundo lugar.
El segundo y tercer lugar, Kevin Durant (23,233) y James Harden (23,136), est√°n muy cerca entre s√≠.

### üìà Tendencias de puntuaci√≥n mensual

```python
# Convertir ambas fechas a tipo datetime
play_by_play_clean['game_date'] = pd.to_datetime(play_by_play_clean['game_date'])
dim_tiempo['fecha'] = pd.to_datetime(dim_tiempo['fecha'])

#union por fecha
pbp_tiempo = play_by_play_clean.merge(dim_tiempo[['fecha', 'mes']], left_on='game_date', right_on='fecha', how='left')

#Calcular promedio de puntuaci√≥n mensual
# Eliminar espacios y separar en dos columnas
pbp_tiempo[['score_home', 'score_away']] = pbp_tiempo['score'].str.replace(' ', '').str.split('-', expand=True)

# Convertir a num√©rico
pbp_tiempo['score_home'] = pd.to_numeric(pbp_tiempo['score_home'], errors='coerce')
pbp_tiempo['score_away'] = pd.to_numeric(pbp_tiempo['score_away'], errors='coerce')

# Sumar total de puntos del momento
pbp_tiempo['total_score'] = pbp_tiempo['score_home'] + pbp_tiempo['score_away']

# Promedio de puntuaci√≥n por mes
puntos_mes = pbp_tiempo.groupby('mes')['total_score'].mean().reset_index()

#Graficar

sns.lineplot(data=puntos_mes, x='mes', y='total_score', marker='o')
plt.title('üìÜ Tendencia Mensual de Puntos Promedio')
plt.xlabel('Mes')
plt.ylabel('Puntos Promedio')
plt.xticks(range(1, 13))
plt.show()

```

Este gr√°fico muestra c√≥mo var√≠a el puntaje promedio a lo largo de un per√≠odo de 12 meses (eje X, del 1 al 12).
Los puntajes promedio generalmente se mantienen en un rango entre 29 y 30 puntos durante la primera mitad del a√±o (Meses 1-5), con un ligero descenso en el Mes 6 (cercano a 28.5).											
El punto clave es un pico dram√°tico en el Mes 7, donde el promedio de puntos se dispara hasta casi 34 puntos, representando el valor m√°s alto.
A partir del Mes 7, el promedio cae bruscamente hasta el Mes 10, alcanzando su punto m√°s bajo (cercano a 27.5 puntos).
Finalmente, hay una ligera recuperaci√≥n en los Meses 11 y 12.

La conclusi√≥n principal es que el patr√≥n de puntaje promedio presenta una alta estacionalidad y variabilidad.											
El Mes 7 es un valor at√≠pico y el factor m√°s influyente en esta tendencia, indicando que, por alguna raz√≥n espec√≠fica de ese mes (como un evento, un torneo especial, o un cambio en la muestra/jugadores), el promedio de puntos anotados fue significativamente superior al de cualquier otro momento del a√±o.											
                                                                                    
El patr√≥n sugiere que, fuera de ese pico, el rendimiento ofensivo es relativamente estable, pero el Mes 10 representa el momento de menor producci√≥n ofensiva en promedio.

# Comparaci√≥n de equipo local vs visitante en un solo partido

```python
import matplotlib.pyplot as plt

# Ejemplo para un solo juego
game = fact_partidos_2010_2023.iloc[0]

labels = ['Puntos', 'FGM', 'FGA', 'FG%', 'FG3M', 'FG3A', 'FG3%', 'FTM', 'FTA', 'FT%', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF']

home_stats = [
    game['pts_home_x'], game['fgm_home'], game['fga_home'], game['fg_pct_home'],
    game['fg3m_home'], game['fg3a_home'], game['fg3_pct_home'],
    game['ftm_home'], game['fta_home'], game['ft_pct_home'],
    game['reb_home'], game['ast_home'], game['stl_home'], game['blk_home'],
    game['tov_home'], game['pf_home']
]

away_stats = [
    game['pts_away_x'], game['fgm_away'], game['fga_away'], game['fg_pct_away'],
    game['fg3m_away'], game['fg3a_away'], game['fg3_pct_away'],
    game['ftm_away'], game['fta_away'], game['ft_pct_away'],
    game['reb_away'], game['ast_away'], game['stl_away'], game['blk_away'],
    game['tov_away'], game['pf_away']
]

x = range(len(labels))
plt.figure(figsize=(14, 6))
plt.barh(x, home_stats, color='skyblue', label='Local')
plt.barh(x, [-v for v in away_stats], color='lightcoral', label='Visitante')
plt.yticks(x, labels)
plt.axvline(0, color='black')
plt.title('Comparaci√≥n Estad√≠stica - Partido')
plt.legend()
plt.tight_layout()
plt.show()
```

El gr√°fico es una pir√°mide de poblaci√≥n o gr√°fico de barras doble que compara las estad√≠sticas promedio por partido entre el equipo Local (azul claro, derecha) y el equipo Visitante (rojo salm√≥n, izquierda).
Las variables en el eje vertical representan m√©tricas de juego (estad√≠sticas clave en el baloncesto). El eje horizontal indica el valor promedio para cada grupo (Local/Visitante).
