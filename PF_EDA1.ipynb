```python
#calcular puntos por jugador

# Crear copia del DataFrame
pbp = play_by_play_clean.copy()

# Crear columna de puntos por evento
def obtener_puntos(row):
    if row['eventmsgtype'] == 3:      # Tiro libre convertido
        return 1
    elif row['eventmsgtype'] == 1:    # Tiro de campo convertido (asumimos 2)
        return 2
    else:
        return 0

# Aplicar la función
pbp['puntos'] = pbp.apply(obtener_puntos, axis=1)

# Agrupar por nombre del jugador y sumar los puntos
top_jugadores = (
    pbp.groupby('player1_name')['puntos']
    .sum()
    .reset_index()
    .sort_values(by='puntos', ascending=False)
    .head(10)
)

# Visualizar
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=top_jugadores, x='puntos', y='player1_name', palette='rocket')

plt.title('🏀 Top 10 Jugadores con más puntos (eventos)')
plt.xlabel('Puntos')
plt.ylabel('Jugador')

# Mostrar los valores encima de las barras
for i, valor in enumerate(top_jugadores['puntos']):
    ax.text(valor + 1, i, int(valor), color='black', va='center')

plt.tight_layout()
plt.show()
```

Este gráfico muestra los diez jugadores con la mayor cantidad de puntos acumulados (o eventos de puntuación) en el conjunto de datos analizado.
LeBron James lidera la lista de forma destacada, con 27,057 puntos, estableciendo una clara diferencia sobre el segundo lugar.
El segundo y tercer lugar, Kevin Durant (23,233) y James Harden (23,136), están muy cerca entre sí.

### 📈 Tendencias de puntuación mensual

```python
# Convertir ambas fechas a tipo datetime
play_by_play_clean['game_date'] = pd.to_datetime(play_by_play_clean['game_date'])
dim_tiempo['fecha'] = pd.to_datetime(dim_tiempo['fecha'])

#union por fecha
pbp_tiempo = play_by_play_clean.merge(dim_tiempo[['fecha', 'mes']], left_on='game_date', right_on='fecha', how='left')

#Calcular promedio de puntuación mensual
# Eliminar espacios y separar en dos columnas
pbp_tiempo[['score_home', 'score_away']] = pbp_tiempo['score'].str.replace(' ', '').str.split('-', expand=True)

# Convertir a numérico
pbp_tiempo['score_home'] = pd.to_numeric(pbp_tiempo['score_home'], errors='coerce')
pbp_tiempo['score_away'] = pd.to_numeric(pbp_tiempo['score_away'], errors='coerce')

# Sumar total de puntos del momento
pbp_tiempo['total_score'] = pbp_tiempo['score_home'] + pbp_tiempo['score_away']

# Promedio de puntuación por mes
puntos_mes = pbp_tiempo.groupby('mes')['total_score'].mean().reset_index()

#Graficar

sns.lineplot(data=puntos_mes, x='mes', y='total_score', marker='o')
plt.title('📆 Tendencia Mensual de Puntos Promedio')
plt.xlabel('Mes')
plt.ylabel('Puntos Promedio')
plt.xticks(range(1, 13))
plt.show()

```

Este gráfico muestra cómo varía el puntaje promedio a lo largo de un período de 12 meses (eje X, del 1 al 12).
Los puntajes promedio generalmente se mantienen en un rango entre 29 y 30 puntos durante la primera mitad del año (Meses 1-5), con un ligero descenso en el Mes 6 (cercano a 28.5).											
El punto clave es un pico dramático en el Mes 7, donde el promedio de puntos se dispara hasta casi 34 puntos, representando el valor más alto.
A partir del Mes 7, el promedio cae bruscamente hasta el Mes 10, alcanzando su punto más bajo (cercano a 27.5 puntos).
Finalmente, hay una ligera recuperación en los Meses 11 y 12.

La conclusión principal es que el patrón de puntaje promedio presenta una alta estacionalidad y variabilidad.											
El Mes 7 es un valor atípico y el factor más influyente en esta tendencia, indicando que, por alguna razón específica de ese mes (como un evento, un torneo especial, o un cambio en la muestra/jugadores), el promedio de puntos anotados fue significativamente superior al de cualquier otro momento del año.											
                                                                                    
El patrón sugiere que, fuera de ese pico, el rendimiento ofensivo es relativamente estable, pero el Mes 10 representa el momento de menor producción ofensiva en promedio.

Existe una correlación positiva muy fuerte con fgm_home (Canastas encestadas, 0.85). Esto es lógico, pues más canastas implican más puntos.

Existe una correlación positiva baja con ftm_home (Tiros libres encestados, 0.24) y fta_home (Tiros libres intentados, 0.18).

Entre Estadísticas de Tiro Libre (ftm_home y fta_home): Tienen una correlación extremadamente alta de 0.93, lo que indica que el número de tiros libres encestados está casi directamente relacionado con el número de tiros libres intentados.												
                                                

La conclusión principal es que la efectividad en el tiro (especialmente las canastas de campo encestadas, fgm_home) y la capacidad de distribuir el balón (asistencias, ast_home) son los factores más determinantes para la puntuación total del equipo local (pts\_home\_x).												
                                                
Aunque intentar tiros (fga_home, fg3a_home) contribuye a la correlación, la conversión de esos intentos (encestar, fgm_home y fg3m_home) tiene un impacto mucho más fuerte y directo en el puntaje final del equipo.												
                                                


# Comparación de equipo local vs visitante en un solo partido

```python
import matplotlib.pyplot as plt

# Ejemplo para un solo juego
game = fact_partidos_2010_2023.iloc[0]

labels = ['Puntos', 'FGM', 'FGA', 'FG%', 'FG3M', 'FG3A', 'FG3%', 'FTM', 'FTA', 'FT%', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF']

home_stats = [
    game['pts_home_x'], game['fgm_home'], game['fga_home'], game['fg_pct_home'],
    game['fg3m_home'], game['fg3a_home'], game['fg3_pct_home'],
    game['ftm_home'], game['fta_home'], game['ft_pct_home'],
    game['reb_home'], game['ast_home'], game['stl_home'], game['blk_home'],
    game['tov_home'], game['pf_home']
]

away_stats = [
    game['pts_away_x'], game['fgm_away'], game['fga_away'], game['fg_pct_away'],
    game['fg3m_away'], game['fg3a_away'], game['fg3_pct_away'],
    game['ftm_away'], game['fta_away'], game['ft_pct_away'],
    game['reb_away'], game['ast_away'], game['stl_away'], game['blk_away'],
    game['tov_away'], game['pf_away']
]

x = range(len(labels))
plt.figure(figsize=(14, 6))
plt.barh(x, home_stats, color='skyblue', label='Local')
plt.barh(x, [-v for v in away_stats], color='lightcoral', label='Visitante')
plt.yticks(x, labels)
plt.axvline(0, color='black')
plt.title('Comparación Estadística - Partido')
plt.legend()
plt.tight_layout()
plt.show()
```

El gráfico es una pirámide de población o gráfico de barras doble que compara las estadísticas promedio por partido entre el equipo Local (azul claro, derecha) y el equipo Visitante (rojo salmón, izquierda).
Las variables en el eje vertical representan métricas de juego (estadísticas clave en el baloncesto). El eje horizontal indica el valor promedio para cada grupo (Local/Visitante).

Variable	    Nombre Completo (Estimado)	            Resultado Clave (Comparación)

Puntos	        Puntos totales	                        El equipo Local anota significativamente más puntos que el Visitante.

FGM	            Canastas de Campo Encestadas	        El equipo Local encesta más canastas de campo.

FGA	            Canastas de Campo Intentadas	        El equipo Local intenta muchas más canastas que el Visitante.

FG%	            Porcentaje de Canastas de Campo	        Son prácticamente iguales, ligeramente a favor del Local.

FG3M	        Triples Encestados	                    El equipo Local encesta más triples.

FG3A	        Triples Intentados	                    El equipo Local intenta más triples.

FTM	            Tiros Libres Encestados	                El equipo Local encesta más tiros libres.

FTA	            Tiros Libres Intentados	                El equipo Local intenta más tiros libres.

FT%	            Porcentaje de Tiros Libres	            Son prácticamente iguales, ligeramente a favor del Local.

REB	            Rebotes	                                El equipo Local captura significativamente más rebotes.

AST	            Asistencias	                            El equipo Local tiene más asistencias.

STL	            Robos	                                Son prácticamente iguales, ligeramente a favor del Local.

BLK	            Tapones	                                Son prácticamente iguales, ligeramente a favor del Local.

TOV	            Pérdidas de Balón (Turnovers)	        El equipo Local tiene menos pérdidas (Visitante tiene más).

PF	            Faltas Personales	                    El equipo Visitante comete más faltas.

La conclusión fundamental es que el equipo local tiene una clara y fuerte ventaja estadística en casi todas las categorías ofensivas y defensivas por partido.											
                                            
El equipo local no solo anota más puntos (la variable más grande del gráfico), sino que lo logra a través de:											
                                            
1. Mayor Volumen Ofensivo: Intentan y encestan consistentemente más canastas de campo, más triples y más tiros libres.											
                                            
2. Mejor Cuidado del Balón: Tienen menos pérdidas de balón (TOV).											
                                            
3. Dominio Físico: Capturan significativamente más rebotes (REB).											
                                            
Esto subraya el concepto de ventaja de localía, indicando que jugar en casa no solo afecta la moral o el entorno, sino que se traduce directamente en un rendimiento estadístico superior en el campo de juego. El equipo visitante, en contraste, se desempeña consistentemente peor en la mayoría de las métricas clave.
